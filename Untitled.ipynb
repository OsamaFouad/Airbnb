```python
guests    = []
bedrooms  = []
numOfbeds = []
baths     = []


linkNum = 1
for link in links:
    airbnb = requests.get(link)
    content = airbnb.content
    # Parse a content
    soup = BeautifulSoup(content, 'lxml')
    
    guests.append(soup.find_all('ol',{'class':"lgx66tx dir dir-ltr"})[0].find_all('li',{'class':'l7n4lsf dir dir-ltr'})[0].text.replace('·',''))
    bedrooms.append(soup.find_all('ol',{'class':"lgx66tx dir dir-ltr"})[0].find_all('li',{'class':'l7n4lsf dir dir-ltr'})[1].text.replace('·',''))
    numOfbeds.append(soup.find_all('ol',{'class':"lgx66tx dir dir-ltr"})[0].find_all('li',{'class':'l7n4lsf dir dir-ltr'})[2].text.replace('·',''))
    if linkNum != 98:
        
        baths.append(soup.find_all('ol',{'class':"lgx66tx dir dir-ltr"})[0].find_all('li',{'class':'l7n4lsf dir dir-ltr'})[3].text.replace('·',''))
    
    print('LinkNum', linkNum)
    linkNum += 1
```
